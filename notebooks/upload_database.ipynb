```python
import os
import requests
```

```python
# Define the folder path and API endpoint
folder_path = "sagemaker_documentation"
api_endpoint = "http://localhost:8000/upsert"
```

```python
# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    
    # Check if it's a file
    if os.path.isfile(file_path):
        # Open the file in binary mode
        with open(file_path, "rb") as file:
            # Send the file to the API
            response = requests.post(api_endpoint, files={"file": file})
            
            # Print the response from the server
            print(f"Uploaded {file_name}, Response: {response.json()}")
```
